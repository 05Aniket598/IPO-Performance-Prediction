```python
def Data_cleaning(df):
    return (
        df       
        .assign(
            **{
                col:df[col].str.strip()
                for col in df.select_dtypes(include='O')
            }
        )
        .rename(columns=str.lower)
        .assign(
            year = lambda df_ : pd.to_datetime(df_['year'].str.replace(" '",''),format="%b%y"),
            cash_from_operating_act_ann_cr = lambda df_ : pd.to_numeric(df_['cash from operating act. ann.'].str.replace(",","")),
            ebitda_ann_cr = lambda df_ : pd.to_numeric(df_['ebitda ann.'].str.replace(",","")),
            eps_adj_latest_ann = lambda df_ : pd.to_numeric(df_['eps adj. latest ann.'].str.replace(",","")),
            npm_ann_percent = lambda df_ : pd.to_numeric(df_['npm ann. %'].str.replace(",","").str.replace("%","")),
            net_profit_ann_cr = lambda df_ : pd.to_numeric(df_['net profit ann.'].str.replace(",","")),
            roce_ann_percent = lambda df_ : pd.to_numeric(df_['roce ann. %'].str.replace(",","").str.replace("%","")),
            roe_ann_percent = lambda df_ : pd.to_numeric(df_['roe ann. %'].str.replace(",","").str.replace("%","")),
            total_debt_to_total_equity_ann = lambda df_ : pd.to_numeric(df_['total debt to total equity ann.'].str.replace(",","")),
        )
        .drop(["cash from operating act. ann.","ebitda ann.","eps adj. latest ann.","npm ann. %","net profit ann.","roce ann. %",
            "roe ann. %","total debt to total equity ann.","total rev. ann."
        ],axis=1)
    )

Data_cleaning(trans_df)
```

```python
def conversion(row):
    try:
        if "M" in row:
            row = pd.to_numeric(row.replace("M","")) * 10
            return row
        else:
            row = pd.to_numeric(row.replace("K","")) / 100000
            return row

    except:
        if type(row) == float:
            return row
```

```python
def fill_max(row):
    if pd.isna(row['max_price_range']):
        return row['min_price_range']
    else:
        return row['max_price_range']
```

```python
def calculate_cffoa_change(row):
    """
    Calculates percentage change using a hybrid approach based on our discussion.
    
    Args:
        row (pandas.Series): A row from the DataFrame containing financial data.
    
    Returns:
        float: The calculated change score.
    """
    start_value = row['cash_from_operating_act_ann_cr_1']
    end_value = row['cash_from_operating_act_ann_cr_3']
    
    # Scenario 1: Handle the zero-division case first.
    if start_value == 0:
        # If the second year is also zero, the change is the third year's value.
        if row['cash_from_operating_act_ann_cr_2'] == 0:
            return row['cash_from_operating_act_ann_cr_3']
        # Otherwise, calculate the change from the second year.
        else:
            return ((row['cash_from_operating_act_ann_cr_3'] - row['cash_from_operating_act_ann_cr_2']) / abs(row['cash_from_operating_act_ann_cr_2'])) * 100

    # Scenario 2: Use the modified percentage change formula for all other cases.
    # Taking the absolute value of the start_value correctly handles
    # changes from negative numbers.
    else:
        return ((end_value - start_value) / abs(start_value)) * 100
```

```python
def calculate_der_change(row):
    """
    Calculates percentage change using a hybrid approach based on our discussion.
    
    Args:
        row (pandas.Series): A row from the DataFrame containing financial data.
    
    Returns:
        float: The calculated change score.
    """
    start_value = row['total_debt_to_total_equity_ann_1']
    end_value = row['total_debt_to_total_equity_ann_3']
    
    # Scenario 1: Handle the zero-division case first.
    if start_value == 0:
        # If the second year is also zero, the change is the third year's value.
        if row['total_debt_to_total_equity_ann_2'] == 0:
            return row['total_debt_to_total_equity_ann_3']
        # Otherwise, calculate the change from the second year.
        else:
            return ((row['total_debt_to_total_equity_ann_3'] - row['total_debt_to_total_equity_ann_2']) / abs(row['total_debt_to_total_equity_ann_2'])) * 100

    # Scenario 2: Use the modified percentage change formula for all other cases.
    # Taking the absolute value of the start_value correctly handles
    # changes from negative numbers.
    else:
        return ((end_value - start_value) / abs(start_value)) * 100
```

```python
def calculate_cagr(row):
    current_gain_percent = row['current gain %']
    year = pd.to_datetime(row['listing date']).year
    years = 2025 - year
    
    # Check for valid gain percentage
    if pd.isna(current_gain_percent):
        return np.nan
    
    # Formula: ((1 + Total Gain)^(1/Years)) - 1
    # Convert gain percentage to a decimal
    total_gain_factor = 1 + (current_gain_percent / 100)
    
    # Check for invalid values that could lead to complex numbers
    if total_gain_factor <= 0:
        return np.nan

    # Calculate and return the CAGR
    return round(((total_gain_factor ** (1 / years)) - 1) * 100,4)
```

```python
def calculate_change(df):
    return (
        df
        .assign(
            change_in_cffoa = lambda df_: round(df_.apply(calculate_cffoa_change,axis=1),4),
            change_in_ebitda = lambda df_: round(df_.apply(calculate_ebitda_change,axis=1),4),
            change_in_eps = lambda df_: round(df_.apply(calculate_eps_change,axis=1),4),
            change_in_net_profit = lambda df_: round(df_.apply(calculate_net_profit_change,axis=1),4),
            change_in_npm_percent = lambda df_: round(df_.apply(calculate_npm_change,axis=1),4),
            change_in_roce_percent = lambda df_: round(df_.apply(calculate_roce_change,axis=1),4),
            change_in_roe_percent = lambda df_: round(df_.apply(calculate_roe_change,axis=1),4),
            change_in_debt_to_equity_ratio = lambda df_: round(df_.apply(calculate_der_change,axis=1),4),
            CAGR = lambda df_ : df_.apply(calculate_cagr,axis=1)
        )
        .drop(columns=['cash_from_operating_act_ann_cr_1',
       'cash_from_operating_act_ann_cr_2', 'cash_from_operating_act_ann_cr_3',
       'ebitda_ann_cr_1', 'ebitda_ann_cr_2', 'ebitda_ann_cr_3',
       'eps_adj_latest_ann_1', 'eps_adj_latest_ann_2', 'eps_adj_latest_ann_3',
       'npm_ann_percent_1', 'npm_ann_percent_2', 'npm_ann_percent_3',
       'net_profit_ann_cr_1', 'net_profit_ann_cr_2', 'net_profit_ann_cr_3',
       'roce_ann_percent_1', 'roce_ann_percent_2', 'roce_ann_percent_3',
       'roe_ann_percent_1', 'roe_ann_percent_2', 'roe_ann_percent_3',
       'total_debt_to_total_equity_ann_1', 'total_debt_to_total_equity_ann_2',
       'total_debt_to_total_equity_ann_3'])
    )
    
```
