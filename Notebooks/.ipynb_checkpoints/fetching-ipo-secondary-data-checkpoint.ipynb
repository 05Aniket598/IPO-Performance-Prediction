```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
import time
import numpy as np
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = 'https://trendlyne.com/ipo/screener/year-2025/'

data = []
def wait_to_load(driver,wait):
    try:
        wait.until(
            lambda d:d.execute_script('return document.readyState') == 'complete'
        )
    except:
        time.sleep(1)

wait = WebDriverWait(driver,10)

for Year in range(2019,2020):

    driver.get(f'https://trendlyne.com/ipo/screener/year-{Year}/')  
    wait_to_load(driver,wait)
    wait.until(
        ec.presence_of_element_located((By.XPATH,'//*[@id="ipo-section"]/main/div[2]/div[2]/div[1]/table'))
    )

    select_feild = wait.until(
        ec.element_to_be_clickable((By.XPATH,'//*[@id="ipo-section"]/main/div[2]/div[2]/div[2]/div/div/select'))
    )

    select_button = Select(select_feild)
    select_button.select_by_visible_text('All')

    last_count = 0
    stable_count = 0
    
    for _ in range(20):  # check up to 20 times (i.e., max 20 seconds)
        rows = driver.find_elements(By.XPATH, '//*[@id="ipo-section"]/main/div[2]/div[2]/div[1]/table/tbody/tr')
        current_count = len(rows)
        
        if current_count == last_count:
            stable_count += 1
            if stable_count >= 10:  # count stable 6 times in a row
                break
        else:
            stable_count = 0
            last_count = current_count
        
        time.sleep(1)  # wait 1 second before checking again
    
    company_count = len(rows)
    print(f"{Year}: {company_count} companies loaded.")


    company_counter = 0
    
    for row in rows:
        # time.sleep(2)
        wait_to_load(driver,wait)
        values = row.find_element(By.TAG_NAME,'td')

        success = False
        for _ in range(10):
            try:
                wait.until(
                    ec.element_to_be_clickable((By.XPATH,'//*[@id="ipo-section"]/main/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[1]/a'))
                )
                link = values.find_element(By.TAG_NAME, 'a')
                company_name = values.text.strip()

                success = True
                break

            except Exception as e:
                print(f"[Retryng stale issue] {e}")
                time.sleep(1)

        if not success:
            print(f"Skipping row {count}")
            company_counter += 1
            continue

        time.sleep(1)
        # link.send_keys(Keys.PAGE_UP)
        link.click()
        wait_to_load(driver,wait)
        # print('Link Clicked')

        Subscription_loaded = False
        for _ in range(5): # Check 5 times
            try:
                wait.until(
                    ec.presence_of_element_located((By.XPATH,'//*[@id="ipo-section"]/main/section[1]/div[1]/table[2]/tbody/tr'))
                )
                subscription_status_feild = driver.find_element(By.XPATH,'//*[@id="ipo-section"]/main/section[1]/div[1]/table[2]/tbody/tr')
                subscription_status = subscription_status_feild.find_element(By.TAG_NAME,'td').text.strip()
                if '-' == subscription_status:
                    raise ValueError('Instead of value hypen(-) is present.')
                break
            except Exception as e:
                print(f'Re Checking the presence of subscritpion status loaded or not. Error : {e}')
                time.sleep(1)

        
        if not Subscription_loaded:
            print('Subscription status not available')
            subscription_status = np.nan


        secondary_data_available = False
        for _ in range(5):
            try:
                wait.until(
                    ec.presence_of_element_located((By.XPATH,'//*[@id="ipo-section"]/main/section[1]/div[2]/table'))
                )
                secondary_data_rows = driver.find_element(By.XPATH,'//*[@id="ipo-section"]/main/section[1]/div[2]/table/tbody/tr')
                secondary_data_rows_value = secondary_data_rows.find_elements(By.TAG_NAME,'td')
                secondary_data_available = True
                break

            except Exception as e:
                print(f'Re Checking the presence of secondary data in table. Error : {e} ')
                time.sleep(1)

        minimum_investment = secondary_data_rows_value[0].text.strip()
        lot_size = secondary_data_rows_value[1].text.strip()
        no_of_shares = secondary_data_rows_value[2].text.strip()
        price_range = secondary_data_rows_value[3].text.strip()
        post_issue_promoter_holding = secondary_data_rows_value[4].text.strip()
        issue_size = secondary_data_rows_value[5].text.strip()

        headers_available = False
        for _ in range(5):
            try:
                wait.until(
                    ec.presence_of_element_located((By.CLASS_NAME,'ipo-stock-name'))
                )
                further_detail_feild = driver.find_element(By.CLASS_NAME,'ipo-stock-name')
                further_detail = further_detail_feild.find_element(By.TAG_NAME,'a')
                headers_available = True

            except Exception as e:
                print(f'Retrying to find link for further detail . Error : {e}')
                time.sleep(1)

        
        further_detail.click()
        # print('Entered in Company detail page')

        for _ in range(5):
            try:
                sec_ind_field = wait.until(
                    ec.presence_of_element_located((By.CLASS_NAME,'tl_breadcrumb'))
                )
                sec_ind_info = sec_ind_field.find_elements(By.TAG_NAME,'li')
                sector = sec_ind_info[1].text.strip()
                industry = sec_ind_info[2].text.strip()

                if '-' in [sector,industry]:
                    raise ValueError('Instead of value hypen(-) is present.\nRefreshing the page....')

            except Exception as e:
                print(f'Setor Industry info is missing. Error{e}')
                driver.refresh()
                time.sleep(2)

        for _ in range(5):
            try:
                nse_bse = wait.until(
                    ec.presence_of_element_located((By.CLASS_NAME,'stock_info_details'))
                )

                nse_bse = nse_bse.find_element(By.CLASS_NAME,'stock_exchange_details')
                
                nse_symbol_bse_code = nse_bse.text.strip()

                if nse_symbol_bse_code == '-':
                    raise ValueError('Content of page is not loaded fully./nRefreshing the page....')

            except Exception as e:
                print(f'Nse Bse detail is not loaded . Error{e}')
                driver.refresh()
                time.sleep(2)
                


        data.append(
            {
                'Company':company_name,
                'Subscritpion Status':subscription_status,
                'Minimum Investment': minimum_investment,
                'Lot size':lot_size,
                'Issue size':issue_size,
                'No. of Shares':no_of_shares,
                'Price Range':price_range,
                'Post issue promoter holding':post_issue_promoter_holding,
                'Sector':sector,
                'Industry':industry,
                'Nse_symbol_Bse_Code':nse_symbol_bse_code
            }
        )
        company_counter += 1
        
        

        # time.sleep(3)
        driver.back()
        wait_to_load(driver,wait)
        driver.back()

        wait_to_load(driver,wait)
        # break

    print(f"Year {Year} : Total {company_counter} company data scraped(collected).")
                  


time.sleep(2)
driver.quit()
```

```python
df = pd.DataFrame(data)
```
